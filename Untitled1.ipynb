```python
import pandas as pd
import matplotlib.pyplot as plt
import arviz as az
import pymc3 as pm

%config InlineBackend.figure_format = 'retina'
az.style.use("arviz-darkgrid")

import logging
logger = logging.getLogger("pymc3")
logger.propagate = False
from scipy import stats
```

```python
df_enum = pd.read_csv("outputs/v1_ubx/enumeration_lfa_v1.csv")
df_enum[[col for col in df_enum.columns if 'accuracy' in col]] = df_enum[[col for col in df_enum.columns if 'accuracy' in col]] * 20
df_enum['total_resp'] = 20

df_wm = pd.read_csv("outputs/v1_ubx/workingmemory_lfa.csv")
df_wm[[col for col in df_wm.columns if 'accuracy' in col]] = df_wm[[col for col in df_wm.columns if 'accuracy' in col]] * 12
df_wm['total_resp'] = 12

df_mot = pd.read_csv("outputs/v1_ubx/moteval_lfa.csv")
df_mot[[col for col in df_mot.columns if 'accuracy' in col]] = df_mot[[col for col in df_mot.columns if 'accuracy' in col]] * 15
df_mot[[col for col in df_mot.columns if 'accuracy' in col]] = df_mot[[col for col in df_mot.columns if 'accuracy' in col]].apply(lambda x: x.round())
df_mot['total_resp'] = 15

df_tsw = pd.read_csv("outputs/v1_ubx/taskswitch_treatment.csv")
df_tsw = df_tsw[['participant_id', 'condition', 'task_status', 'relative-switch-correct', 'relative-unswitch-correct', 
                 'relative-switch-nb', 'relative-unswitch-nb', 'parity-switch-correct', 'parity-unswitch-correct',
                 'parity-switch-nb', 'parity-unswitch-nb']]

df_lb = pd.read_csv("outputs/v1_ubx/loadblindness_lfa.csv")
df_lb[[col for col in df_lb.columns if 'accuracy' in col]] = df_lb[[col for col in df_lb.columns if 'accuracy' in col]] * 20
df_lb['total_resp'] = 20
df_lb.head()

df_go = pd.read_csv("outputs/v1_ubx/gonogo_lfa.csv")
df_go[[col for col in df_go.columns if 'accuracy' in col]] = df_go[[col for col in df_go.columns if 'accuracy' in col]] * 25
df_go.head()

# tasks = {'enumeration': df_enum, 'working_memory': df_wm,'gonogo': df_go, 'loadblindness': df_lb, 'mot': df_mot}
tasks = {'mot': df_mot}
df_mot.head()
```

```python
def get_pooled_trace(data, condition):
    pre_test = data[data['task_status']=='PRE_TEST']
    post_test = data[data['task_status']=='POST_TEST']
    with pm.Model() as pooled_model: 
        # Independent parameters for each participant
        total_resp = pm.Data("total_resp", pre_test.total_resp.values)
        # prior on theta: 
        pre_test_theta = pm.Uniform("pre_test_theta", lower=0, upper=1)
        post_test_theta = pm.Uniform("post_test_theta", lower=0, upper=1)
        # likelihood
        pre_test_binom = pm.Binomial("pre_test_binomial", p=pre_test_theta, n=total_resp, observed=pre_test[condition])
        post_test_binom = pm.Binomial("post_test_binomial", p=post_test_theta, n=total_resp, observed=post_test[condition])
        diff_of_means = pm.Deterministic("difference_of_means", post_test_theta - pre_test_theta)
        unpooled_trace = pm.sample(2000, return_inferencedata=True)
    return unpooled_trace
```

```python
all_traces_zpdes = {}
all_traces_baseline = {}
for name, task in tasks.items():
    all_traces_zpdes[name] = {}
    all_traces_baseline[name] = {}
    conditions = task[[col for col in task.columns if 'accuracy' in col]]
    for condition in conditions:
        all_traces_zpdes[name][condition] = get_pooled_trace(task[task['condition']=='zpdes'], condition)
        all_traces_baseline[name][condition] = get_pooled_trace(task[task['condition']=='baseline'], condition)
```

```python
# all_traces_zpdes['enumeration']['1-accuracy']['posterior']['theta']
for condition in all_traces_zpdes['mot'].keys():
    az.plot_posterior(
        all_traces_zpdes['mot'][condition],
        color="#87ceeb",
    );
```

```python
for condition in all_traces_baseline['mot'].keys():
    az.plot_posterior(
        all_traces_baseline['mot'][condition],
        color="#87ceeb",
    );
```
